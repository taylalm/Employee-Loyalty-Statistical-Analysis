## Preparing of New Columns and Variables

```python
# Create age group
position_level_mapping = {
    1: ['Accounts Intern', 'Software Engineer I', 'Marketing Intern', 'Marketing Development Representative', 
        'HR Representative', 'QA Engineer I', 'Business Development Representative', 'HR Intern', 
        'Sales Development Representative', 'Account Representative'],
    2: ['Account Associate', 'HR Executive', 'Software Engineer II', 'HR Associate', 'Marketing Representative', 
        'QA Engineer II', 'Account Executive', 'Sales Executive', 'Marketing Associate', 'Sales Representative'],
    3: ['Regional Marketing Manager', 'Senior Marketing Executive', 'QA Lead', 'ERP Head', 'Regional Sales Manager', 
        'Senior Executive', 'Regional Account Head', 'Software Engineer III', 'Senior HR', 'Senior Account Executive'],
    4: ['National Marketing Manager', 'National Sales Manager', 'National Account Head', 'Account Director', 
        'HR Director', 'Sales Director', 'Technical Lead', 'Recruitment Manager', 'Marketing Director', 'IT Manager']
}
position_to_level = {position: level for level, positions in position_level_mapping.items() for position in positions}
employee_df['Position Level'] = employee_df['Position'].map(position_to_level)
employee_df.head()

# Create Insurance Encoding
employee_df['Medical Insurance'] = employee_df['Insurance'].apply(lambda x: 1 if 'Medical' in str(x) or 'Both' in str(x) else 0)
employee_df['Life Insurance'] = employee_df['Insurance'].apply(lambda x: 1 if 'Life' in str(x) or 'Both' in str(x) else 0)

# Create Employee Status
employee_df['Tenure Before Joining'] = employee_df.apply(lambda row: max(row['Year of Experience'] - row['In Company Years'], 0), axis=1)
employee_df['Age of Joining'] = employee_df.apply(lambda row: max(row['Age'] - row['In Company Years'], 21), axis=1)
employee_df['Long Term Employee'] = employee_df['Tenure Before Joining'].apply(lambda x: 1 if x == 0 else 0)
employee_df['First Role?'] = employee_df.apply(lambda row: 1 if row['Position Level'] == 1 or row['Age of Joining'] < employee_df[employee_df['Position Level'] == row['Position Level']]['Age'].min() else 0, axis=1)
```

## Inference Analysis on Career Progression
Our Team would like to investigate as to why the promotion rate is low and wanted to investigate if externally sourced employees bring any advantage which may result in the low numbers. As such, we conduct a check on the Confidence Interval for Years of Experience between those who are Internally Promoted (First Role?=0) and those Externally Joined (First Role?=1)

### 1. Confidence Interval of Years of Experience for all employees

```python
# Group by Position Level and First Role?, then calculate the mean and count of Year of Experience
experience_stats = employee_df.groupby(['Position Level', 'First Role?'])['Year of Experience'].agg(['mean', 'count']).reset_index()

experience_stats.columns = ['Position Level', 'First Role?', 'Mean Years of Experience', 'Sample Size']

experience_stats_pivot = experience_stats.pivot(index='Position Level', columns='First Role?', values=['Mean Years of Experience', 'Sample Size'])

experience_stats_pivot.columns = [f'{stat} (First Role?={role})' for stat, role in experience_stats_pivot.columns]

experience_stats_pivot.reset_index(inplace=True)

experience_stats_pivot
```

### 2. Confidence Interval of Years of Experience for Internally Promoted Employees (First Role?=0)

```python
from scipy.stats import t

confidence_intervals = employee_df[employee_df['First Role?'] == 0].groupby('Position Level')['Year of Experience'].apply(
    lambda x: t.interval(0.90, len(x)-1, loc=np.mean(x), scale=np.std(x, ddof=1)/np.sqrt(len(x)))
).reset_index()

confidence_intervals.columns = ['Position Level', 'Year of Experience Confidence Interval']
confidence_intervals
```

### 3. Confidence Interval of Years of Experience for Externally Sourced Employees (First Role?=1)

```python
from scipy.stats import t

confidence_intervals = employee_df[employee_df['First Role?'] == 1].groupby('Position Level')['Year of Experience'].apply(
    lambda x: t.interval(0.90, len(x)-1, loc=np.mean(x), scale=np.std(x, ddof=1)/np.sqrt(len(x)))
).reset_index()

confidence_intervals.columns = ['Position Level', 'Year of Experience Confidence Interval']
confidence_intervals
```

### Hypothesis on Career Progression
With the 2 confidence intervals, we want to test if Internally Promoted Employees have longer professional experience than Externally Sourced Employees

H0: μ(Years Of Experience of Internally Promoted Employee) ≤ μ(Years of Experience of Externally Sourced Employee)
H1: μ(Years Of Experience of Internally Promoted Employee) > μ(Years of Experience of Externally Sourced Employee)

At 90% Confidence Level, reject Null Hypothesis if p ≤ 0.10

```python
from scipy.stats import ttest_ind

# Initialize a list to store the results
t_test_results = []

for level in employee_df['Position Level'].unique():
    # Separate the data into two groups based on 'First Role?' for the current position level
    group_0 = employee_df[(employee_df['Position Level'] == level) & (employee_df['First Role?'] == 0)]['Year of Experience']
    group_1 = employee_df[(employee_df['Position Level'] == level) & (employee_df['First Role?'] == 1)]['Year of Experience']
    
    # Perform the one-tailed t-test
    t_stat, p_value = ttest_ind(group_0, group_1, alternative='greater', equal_var=False)
    
    # Store the results
    t_test_results.append({
        'Position Level': level,
        'T-statistic': t_stat,
        'P-value': p_value,
        'Mean Difference': group_0.mean() - group_1.mean()
    })

t_test_results_df = pd.DataFrame(t_test_results)
t_test_results_df.sort_values('Position Level')
```
